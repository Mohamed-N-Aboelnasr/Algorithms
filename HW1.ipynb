# `Q5)`

```python
# Time complexity before improving the implementation -----------> O(n^2)
# improving the implementation
def minDistance(lis : list):
	sortedlis=sorted(lis)
	minDist=sortedlis[len(sortedlis)-1]
	for i in range(len(sortedlis)-1):
		minDist=min(minDist,abs(sortedlis[i+1]-sortedlis[i]))
	return minDist
mylist=[1,5,11,17,25,29,31,39,45,60,64]
print(minDistance(mylist))
# Time complexity of this implementaion -------------> O(nlogn)
```

```python
def findCommon(lis1:list ,lis2:list):
	m=len(lis1)
	n=len(lis2)
	x=0;y=0
	z=0
	while(x<m and y<n):
		if lis1[x]==lis2[y]:
			print(lis1[x],end=' ')
			x+=1
			y+=1
			z+=1
		elif lis1[x]<lis2[y]:
			x+=1
		else:
			y+=1
	print('\n'+'no. of comparisons=',(m+n-z))

list1=[2,5,5,5]
list2=[2,2,3,5,5,7]
findCommon(list1,list2)
# Time complexity -------------------> O(n)
# no. of comparisons=(m + n - no. of common elements)
# maximum no. of comparisons=(m + n) if there is no common elements
```

# `Q7)`

```python
# Euclid's algorithm
# Time complexity -------------------> O(log n)
def gcd(a,b):
	c=0
	while b!=0:
		r=a%b
		a=b
		b=r
		c+=1
	return a,c
print('gcd= ',gcd(105,350)[0])
print('steps are around : ',gcd(105,350)[1])

#Another solution using recursion
#def gcd(a,b):
#	return a if b==0 else gcd(b,a%b)
#print(gcd(31415,14142))

print('######################################################################')

#consecutive integers
# Time complexity -------------------> O(n)
def gcd2(a,b):
	maxCommon=min(a,b)
	c=0
	while a%maxCommon!=0 or b%maxCommon!=0:
		if a%maxCommon==0 and b%maxCommon==0:
			c+=1
			return maxCommon
		else:
			maxCommon-=1
			c+=1
	return maxCommon,c
print('gcd= ',gcd2(105,350)[0])
print('steps are around : ',gcd2(105,350)[1])
print('so here euclids algorithm faster than consequtive algorithm about ',gcd2(105,350)[1]//gcd(105,350)[1],'times')
```

# Factorial Iterative


```python
# Time complexity -------------------> O(n)
def iterativeFactorial(n):
    if n==0:
        return 1
    factorial=1
    for i in range(2,n+1):
        factorial*=i
    return factorial
print(iterativeFactorial(6))
```

# Recursive Factorial


```python
# Time complexity -------------------> O(n)
def recursiveFactorial(n):
    return 1 if n==0 else n*recursiveFactorial(n-1)
print(recursiveFactorial(6))
```

# Fibonacci Iterative


```python
# Time complexity -------------------> O(n)
def fibIterative(n):
    x,y=0,1
    for i in range(n):
        x,y=y,x+y
    return x
print(fibIterative(10))
```

# Fibonacci Recursion

```python
# Time complexity -------------------> O(2^n)
def fibRecursion(n):
    if n<=1:
        return n
    else:
        return fibRecursion(n-1)+fibRecursion(n-2)
print(fibRecursion(10))
```

# Iterative Linear Search

```python
# Time complexity -------------------> O(n)
def iterLinearSearch(arr,num):
    for i in range(len(arr)):
        if arr[i]==num:
            return 'The number was found in index '+str(i)
    return 'The number wasn\'t found in the array'
myarray=[5,8,11,12,17,23,40,49]
number=23
print(iterLinearSearch(myarray,number))
```

# Recursion linear search

```python
# Time complexity -------------------> O(n)
def recLinearSearch(arr,num,index=0):
    if index==len(arr):
        return 'The number wasn\'t found in the array'
    elif arr[index]==num:
        return 'The number was found in index '+str(index)
    else: return recLinearSearch(arr,num,index+1)
myarray=[5,8,11,12,17,23,40,49]
number=23
print(recLinearSearch(myarray,number))
```

# Binary Search Iterative

```python
# Time complexity -------------------> O(log n)
def iterBinarySearch(arr,num):
    low=0; high=len(arr)-1
    while low<=high:
        mid=(low+high)//2
        if num==arr[mid]:
            return 'The number was found in index '+str(mid)
        elif num<arr[mid]:
            high=mid-1
        else:
            low=mid+1
    return 'The number wasn\'t found in the array'
myarray=[5,8,11,12,17,23,40,49]
number=23
print(iterBinarySearch(myarray,number))
```

# Binary Search Recursive

```python
# Time complexity -------------------> O(log n)
def recBinarySearch(arr,num,start,end):
    low=start; high=end
    mid=(low+high)//2
    if low>high:
        return 'The number wasn\'t found in the array'
    elif num==arr[mid]:
        return 'The number was found in index '+str(mid)
    elif num<arr[mid]:
        return recBinarySearch(arr,num,low,end=mid-1)
    else:
        return recBinarySearch(arr,num,start=mid+1,end=high)
myarray=[5,8,11,12,17,23,40,49]
number=23; start=0; end=len(myarray)-1
print(recBinarySearch(myarray,number,start,end))
```
